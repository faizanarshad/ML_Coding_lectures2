#### Normalizing the data 

```python
ks4 = kashti[["age in days",'fare']]
ks4.head()
```

# Method of Normalizzation
    - simple feature scaling
    - min max method
    - z score(standard score) -3 -to- +3
    - Log Transformation

```python
# simple features scaling
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4['age in days'] = ks4['age in days']/ks4['age in days'].max()
ks4.head()
```

```python
ks4.head()
```

```python
# min - Max method
ks4['fare'] = (ks4['fare']- ks4['fare'].min()) / ( ks4['fare'].max()- ks4['fare'].min())
ks4.head()
```

```python
# Z score (standard score)
ks4['fare'] = (ks4['fare']- ks4['fare'].mean())/ks4['fare'].std()
ks4.head()
```

## Log Transformation

```python
ks4['fare'] = np.log(ks4['fare'])
ks4.head()
```

# Binning 
- group of values into smaller number of values(bins)
- conver numaric into categories(jawan,boorhay,Bechay)or 1-16,17-30 etc
- to have better understanding of groups
- 
        - lov vs mid vs high price
    

```python
bins = np.linespace(min(kashti['age']),max(kashti['age'],15000))
age_groups = ["bachay","Jawan","Boorhay"]
kashti['age'] = pd.cut(kashti['age'],bins,labels=age_groups,include_lowest=True)
kashti['age']
```

```python
df['ageD'], bins = pd.qcut(df.iloc[:, 3], 2, retbins=True)
left = (df["age"].values <= bins[:,None]).T.astype(int)
dl = pd.DataFrame(left, columns=["(-inf, {}]".format(b) for b in bins])
dr = pd.DataFrame(1-left, columns=["({}, +inf)".format(b) for b in bins])
dout = pd.concat([pd.get_dummies(df), dl, dr], axis=1)
dout
```
